```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import pandas as pd
import numpy as np

import statsmodels.api as sm1 #Usa OLS com X como matriz de planejamento
import statsmodels.formula.api as sm  #Usa OLS como no R

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
dados = pd.read_excel('../entrega1/Datasets/dados_preparados.xls')
```

```python
dados.describe()
```

```python
f, ax = plt.subplots(2,2,figsize=(16,8))

ax[0,0].scatter(dados.CALORIES_INTAKE, dados.MORTALIDADE, c='green', alpha=0.8)  # Plot the raw data
ax[0,0].set_xlabel("Alimentação(Kilocalorias / pessoa e dia)")
ax[0,0].set_ylabel("Mortalidade Infantil "'\n'"(entre 0 e 5 anos de idade"'\n'"mortos a cada 1,000 nascidos)")

ax[0,1].scatter(dados.POBREZA, dados.MORTALIDADE, c='red', alpha=0.8)  # Plot the raw data
ax[0,1].set_xlabel("Pobreza (% de pessoas vivendo com menos de U$ 2 por dia)")
ax[0,1].set_ylabel("Mortalidade Infantil "'\n'"(entre 0 e 5 anos de idade"'\n'"mortos a cada 1,000 nascidos)")

ax[1,0].scatter(np.log(dados.CALORIES_INTAKE), dados.MORTALIDADE, c='green', alpha=0.8)  # Plot the raw data
ax[1,0].set_xlabel("ln(Alimentação(Kilocalorias / pessoa e dia)")
ax[1,0].set_ylabel("Mortalidade Infantil "'\n'"(entre 0 e 5 anos de idade"'\n'"mortos a cada 1,000 nascidos)")

ax[1,1].scatter(np.log(dados.POBREZA), dados.MORTALIDADE, c='red', alpha=0.8)  # Plot the raw data
ax[1,1].set_xlabel("ln(Pobreza (% de pessoas vivendo com menos de U$ 2 por dia))")
ax[1,1].set_ylabel("Mortalidade Infantil "'\n'"(entre 0 e 5 anos de idade"'\n'"mortos a cada 1,000 nascidos)")

plt.tight_layout(1)
```
